*** part two

```python
peo=["swetha","lucky","ammu","chinnu","minnu"]
age=[19,15,12,18,13]
weight=[55,45,45,25,50]
height=[109,135,133,111,173]
print(peo)
print(age)
print(weight)
print(height)
```

```python
import matplotlib.pyplot as plt
plt.scatter(weight , height)
plt.ylabel("height")
plt.xlabel("weight")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.scatter(age , height)
plt.ylabel("height")
plt.xlabel("age")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.scatter(weight , age)
plt.ylabel("age")
plt.xlabel("weight")
plt.show()
```

bar plot

```python
plt.figure(figsize=(5,3))
plt.title("people weight in kg", fontsize=16, fontstyle='italic')
plt.bar(x=peo, height=weight, width=0.6, color='yellow', edgecolor='k',alpha=0.6)
plt.xlabel("people",fontsize=15)
plt.ylabel("weight",fontsize=15)
plt.show()
```

histogram

```python
import numpy as np
plt.figure(figsize=(5,5))
plt.hist(weight, color="pink",edgecolor='k',alpha=0.6, bins=5)
plt.title("histogrm",fontsize=15)
plt.xlabel("weight",fontsize=15)
plt.ylabel("weight in scale",fontsize=15)
plt.show()
```
